```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Integer, String, Column, text

# Read the CSV dataa

df = pd.read_csv('../GPT4_setup/corona_India_latest.csv',index_col=0)

# Define your SQLAlchemy engine
# Here we are using sqlite, but you could use other systems like PostgreSQL, MySQL, etc.
engine = create_engine('sqlite:///corona.db')



# Load the data from the DataFrame to the database
df.to_sql('india_corona', engine, if_exists='replace', index=False)
```

```python
with engine.connect() as connection:
    # Define your SQL statement
    sql = text('SELECT * FROM india_corona')

    # Execute the statement
    result = connection.execute(sql)

    # Fetch all rows
    rows = result.fetchall()

    # Print the rows
    for row in rows:
        print(row)
```
