```python
import requests
from bs4 import BeautifulSoup
parent_site='https://www.coingecko.com/en/categories'
# make a request to the website
response = requests.get(parent_site)

# parse the downloaded data
# parse the downloaded data
# parse the downloaded data
soup = BeautifulSoup(response.text, 'html.parser')

# find all tr elements
tr_elements = soup.find_all('tr')

# create a list to hold the links
links = []
# fetch_urls=[]
categories=[]
cat_dict={}
# iterate over each tr element
for tr_element in tr_elements:
    # find all b elements within this tr
    b_elements = tr_element.find_all('b')
    # iterate over each b element
    for b_element in b_elements:
        # find all a elements within this b
        a_elements = b_element.find_all('a')
        categories.append(b_element.text.strip())
        # iterate over each a element
        for a_element in a_elements:
            # get the href attribute of the a element
            href = a_element.get('href').replace('/en/categories','')
            # append the link to the list
            links.append(href)
            







```

```python
def flatten_list(list_of_lists):
  """Flattens a list of lists into a single list.

  Args:
    list_of_lists: A list of lists.

  Returns:
    A single list containing all of the elements from the input lists.
  """

  flat_list = []
  for sublist in list_of_lists:
    flat_list.extend(sublist)
  return flat_list
```

```python
def regex_fetch(st):
    import re

    # regex to find everything after "/categories/"

    pattern = re.compile(r'/categories/(.*)')

    # find the match in the string
    match = pattern.search(st)
    return(match.group(1))
```
